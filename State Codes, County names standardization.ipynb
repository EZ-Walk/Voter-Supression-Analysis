# Problem

the economics table uses lower case county and state names while the demographics and polling uses Titled county names and state codes.

# Solution
1. Convert the columns in economics to match the columns in demsXpolling

```python
econs['State'] = econs['State'].apply(lambda x: x.title()).replace(state_codes)
econs['County'] = econs['County'].apply(lambda x: x.title())
econs.head()
```

```python
dems['State'] = dems['State'].apply(lambda x: x.title()).replace(state_codes)
dems['County'] = dems['County'].apply(lambda x: x.title())
dems.head()
```
